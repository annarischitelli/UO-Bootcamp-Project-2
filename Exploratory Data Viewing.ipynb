```python
yearly_generation = "Resources/annual_generation_state.csv"
yearly_generation_pd = pd.read_csv(yearly_generation, encoding="ISO-8859-1")
yearly_generation_pd = yearly_generation_pd.loc[yearly_generation_pd["Energy Source"]!="Total"]
yearly_generation_pd.replace(',','', regex=True, inplace=True)
yearly_generation_pd["Generation (Megawatthours)"] = yearly_generation_pd["Generation (Megawatthours)"].astype(float)
yearly_generation_pd = yearly_generation_pd.rename(columns = {'ï»¿Year':"Year"})

yearly_generation_pd.to_csv("Resources/annual_generation_state_clean.csv")
yearly_generation_pd.head()
```

```python
yearly_capacity = "Resources/existcapacity_annual (1).csv"
yearly_capacity_pd = pd.read_csv(yearly_capacity, encoding="ISO-8859-1")
yearly_capacity_pd = yearly_capacity_pd.loc[yearly_capacity_pd["Energy Source"]!="All Sources"]
yearly_capacity_pd.replace(',','', regex=True, inplace=True)
yearly_capacity_pd['Generators'] = yearly_capacity_pd['Generators'].replace(np.nan, 0)
yearly_capacity_pd['Facilities'] = yearly_capacity_pd['Facilities'].replace(np.nan, 0)
yearly_capacity_pd = yearly_capacity_pd.rename(columns = {'ï»¿Year':"Year"})

yearly_capacity_pd.to_csv("Resources/existcapacity_annual_clean.csv")
yearly_capacity_pd.head(100)
```
